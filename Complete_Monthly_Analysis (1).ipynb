```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
jan2023_df = pd.read_csv("January2023Status.csv")
feb2023_df = pd.read_csv("February2023Status.csv")
march2023_df = pd.read_csv("March2023Status.csv")
april2023_df = pd.read_csv("April2023Status.csv")
may2023_df = pd.read_csv("May2023Status.csv")
june2023_df = pd.read_csv("June2023Status.csv")
july2023_df = pd.read_csv("July2023Status.csv")
aug2023_df = pd.read_csv("August2023Status.csv")
```

# Monthly Analysis of Underweight Children

```python
def underweightNormal_sum(month_df):
    return month_df['Tot_UW_Normal'].sum(axis = 0)
def MUW_sum(month_df):
    return month_df['Tot_MUW'].sum(axis = 0)
def SUW_sum(month_df):
    return month_df['Tot_SUW'].sum(axis = 0)

UWNormalList = [
    underweightNormal_sum(jan2023_df),
    underweightNormal_sum(feb2023_df),
    underweightNormal_sum(march2023_df),
    underweightNormal_sum(april2023_df),
    underweightNormal_sum(may2023_df),
    underweightNormal_sum(june2023_df),
    underweightNormal_sum(july2023_df),
    underweightNormal_sum(aug2023_df),
]

MUWList = [
    MUW_sum(jan2023_df),
    MUW_sum(feb2023_df),
    MUW_sum(march2023_df),
    MUW_sum(april2023_df),
    MUW_sum(may2023_df),
    MUW_sum(june2023_df),
    MUW_sum(july2023_df),
    MUW_sum(aug2023_df)
]

SUWList = [
    SUW_sum(jan2023_df),
    SUW_sum(feb2023_df),
    SUW_sum(march2023_df),
    SUW_sum(april2023_df),
    SUW_sum(may2023_df),
    SUW_sum(june2023_df),
    SUW_sum(july2023_df),
    SUW_sum(aug2023_df)
]
print(UWNormalList)
print(MUWList)
print(SUWList)
```

```python
months = ["January", "February", "March", "April", "May", "June", "July", "August"]
```

```python
plt.plot(months, UWNormalList, marker = 'o', linestyle = '-', label = 'Normally Underweighted', color = 'orange')
plt.xlabel("2023 Months")
plt.ylabel("Number of Underweight Children under Normal Category")
plt.title("Monthly analysis of Normal Underweight Children")
plt.legend()
plt.show()
```

```python
plt.plot(months, MUWList, marker = 'o', linestyle = '-', label = 'Moderately Underweight', color = 'orange')
plt.xlabel("2023 Months")
plt.ylabel("Underweight Children under Moderate Category")
plt.title("Monthly analysis of Moderate Underweight Children")
plt.legend()
plt.show()
```

```python
plt.plot(months, SUWList, marker = 'o', linestyle = '-', label = 'Severly Underweight', color = 'orange')
plt.xlabel("2023 Months")
plt.ylabel("Underweight Children under Severe Category")
plt.title("Monthly analysis of Severe Underweight Children")

plt.legend()
plt.show()
```

# Bar Graph for Severly Underweight, Wasted and Stunted Children

```python
bar_width = 0.2  # Adjust the bar width as needed
index = np.arange(len(months))



plt.bar(index, wastingSAMList, bar_width, label="Severely Wasted", color='b')
plt.bar(index + bar_width, SUWList, bar_width, label="Severely Underweight", color='g')
plt.bar(index + 2 * bar_width, stuntingSAMList, bar_width, label="Severely Stunted", color='r')

# Add spacing between months by adjusting the x-axis tick positions
plt.xlabel("2023 Months")
plt.ylabel("Number of Children")
plt.title("Monthly Analysis of Severely Wasted, Underweight, and Stunted Children")
plt.xticks(index + bar_width, months)

plt.legend()
plt.tight_layout()  # Ensures that labels and bars fit within the figure

plt.show()
```
